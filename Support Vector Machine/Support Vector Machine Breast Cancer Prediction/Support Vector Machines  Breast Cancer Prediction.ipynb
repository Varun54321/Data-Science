## Check out the data

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Get the Data

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

## Creating Data Frame of features

```python
df_feature=pd.DataFrame(cancer.data,columns=cancer.feature_names)
```

```python
df_feature.head()
```

```python
df_feature.info()
```

```python
df_feature.describe()
```

##   Target Data

```python
cancer.target
```

### Look for missing value in Features dataset

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,7))
sns.heatmap(df_feature.isnull(),yticklabels=False,cbar=False)
plt.show()
```

### * There is no missing values*

## Splitting Training and Testing Data

```python
X=df_feature
y=cancer.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

## Training Support Vector Machine Model

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
svc_predict=svc.predict(X_test)
```

## Predictions and Evaluation of Support Vector Machine

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print('Confusion Matrix :')
print(confusion_matrix(y_test,svc_predict))
print('\n')
print('Classification Report :')
print(classification_report(y_test,svc_predict))

```
